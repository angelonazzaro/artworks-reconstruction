```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
import utils.fourier as ut
```

```python
input_img = cv.imread('5.34.33.png', cv.IMREAD_UNCHANGED)
r, g, b, a = cv.split(input_img)
input_img = cv.merge((r, g, b, a))

input_edge = ut.alpha_edge_detector(input_img)

plt.set_cmap('gray')
plt.imshow(input_edge)
plt.show()
```

```python
edge_pixels_x, edge_pixels_y = np.where(input_edge[:][:] == 1)
edge_pixels_y
```

```python
input_edge[0][24]
```

```python
input_img.shape
```

```python
neighbor_offsets = np.array([[1, 0], [-1, 0], [0, 1], [0, -1]])
rows, cols, channels = input_img.shape
interpolation_region = np.zeros((rows, cols, channels), dtype=np.uint8)
threshold = 5

for i in range(len(edge_pixels_x)):
    x = edge_pixels_x[i]
    y = edge_pixels_y[i]

    for offset_x, offset_y in neighbor_offsets:
        neighbor_x = x + offset_x
        neighbor_y = y + offset_y

        # se siamo fuori oppure il canale alpha Ã¨ 0, ci muoviamo nella direzione opposta
        if (not (0 <= neighbor_x < rows and 0 <= neighbor_y < cols)
                or input_img[neighbor_x][neighbor_y][3] == 0):

            for t in range(1,  threshold + 1): # neighbor = [0, 24] offset = [-1, 0]
                if offset_x > 0 or offset_y > 0:
                    opposite_x = neighbor_x - t if offset_x != 0 else neighbor_x # [0]
                    opposite_y = neighbor_y - t if offset_y != 0 else neighbor_y # [24]
                else:
                    opposite_x = neighbor_x + t if offset_x != 0 else neighbor_x # [0]
                    opposite_y = neighbor_y + t if offset_y != 0 else neighbor_y # [24]

                # controllo se vado di nuovo fuori
                if not (0 <= opposite_x < rows and 0 <= opposite_y < cols):
                    break

                if input_img[opposite_x][opposite_y][3] != 0:
                    interpolation_region[opposite_x][opposite_y] = input_img[opposite_x][opposite_y]
```

```python
plt.imshow(interpolation_region)
plt.show()
```
